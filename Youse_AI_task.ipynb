### ***Data Discription*** 

***CUST_ID:***  Identification of Credit Card holder

***BALANCE:*** Balance amount left in their account to make purchases

***BALANCE_FREQUENCY:*** How frequently the Balance is updated, score between 0 and 1 (1 = frequently updated, 0 = not frequently updated)

***PURCHASES:*** Amount of purchases made from account

***ONEOFF_PURCHASES:*** Maximum purchase amount done in one-go

***INSTALLMENTS_PURCHASES:***  Amount of purchase done in installment

***CASH_ADVANCE:*** Total cash-advance amount

***PURCHASES_ FREQUENCY:*** How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased, 0 = not frequently purchased)

***ONEOFF_PURCHASES_FREQUENCY:*** How frequently Purchases are happening in one-go (1 = frequently purchased, 0 = not frequently purchased)

***PURCHASESINSTALLMENTSFREQUENCY:*** How frequently purchases in installments are being done (1 = frequently done, 0 = not frequently done)

***CASH_ADVANCE_ FREQUENCY:*** How frequently the cash in advance being paid

***CASH_ADVANCE_TRX:*** Number of Transactions made with "Cash in Advanced"

***PURCHASES_TRX:*** Numbe of purchase transactions made

***CREDIT_LIMIT:*** Limit of Credit Card for user

***PAYMENTS:***  Amount of Payment done by user

***MINIMUM_PAYMENTS:***  Minimum amount of payments made by user

***PRC_FULL_PAYMEN:***  Percent of full payment paid by user

***TENURE:*** Number of months as a customer service for user

##### Data is ready for standardization

##### Before standardization let select the features which are important for our analysis and remove non useful or non informative coloums 

```python
imp_feature_col = ['BALANCE_FREQUENCY','CREDIT_LIMIT','Average_purchase_amount','Average_cash_advance','Monthly_purchase','Monthly_cash_advance','Monthly_payments','PRC_FULL_PAYMENT','One_Of_Purchase','Installment_Purchases','Both_the_Purchases','None_Of_the_Purchases','use_for_purchase','use_for_cash_advance','No_use','use_for_both']
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
SS = StandardScaler()

scaled_data = pd.DataFrame(SS.fit_transform(dfc[imp_feature_col]),columns=imp_feature_col)
scaled_data.head()
```

##### Now we will perform PCA 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.85)
```

```python
pca.fit(scaled_data)
```

```python
pca_data = pca.transform(scaled_data)
```

```python
pca_data.shape
```

```python
scaled_data.shape
```

```python
pca_data
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
cluster_range = range(2, 11)  # Test from 2 to 10 clusters
db_scores = []

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    labels = kmeans.fit_predict(pca_data)
    db_index = davies_bouldin_score(pca_data, labels)
    db_scores.append(db_index)
    print(f"Clusters: {n_clusters}, Davies-Bouldin Index: {db_index}")
```

```python
plt.plot(cluster_range, db_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
plt.title('Davies-Bouldin Index vs Number of Clusters')
plt.show()
```

```python
cluster_data = clustered_dfc.groupby('Cluster').apply(lambda x: x[imp_feature_col].mean())
cluster_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
standard_cluster_data_normalized = pd.DataFrame(
    scaler.fit_transform(cluster_data),
    index=cluster_data.index,
    columns=cluster_data.columns
)
```

```python
clusters = standard_cluster_data_normalized.index.unique()
fig, axs = plt.subplots(2, 3, figsize=(24, 12))  # Change to a 2x3 grid
axs = axs.flatten()
feature_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#b5bd61', '#1b9e77', '#d95f02']
features = standard_cluster_data_normalized.columns
y_min = 0
y_max = 1

for i, cluster in enumerate(clusters):
    cluster_data = standard_cluster_data_normalized.loc[cluster]
    cluster_data_T = cluster_data.T
    cluster_data_T.plot.bar(
        ax=axs[i],
        color=feature_colors,
        width=0.8,
        legend=False  # Disable the default legend
    )
    axs[i].set_title(f'Cluster {cluster}', fontsize=16)
    axs[i].set_xlabel('Features', fontsize=10)
    axs[i].set_ylabel('Values', fontsize=14)
    axs[i].tick_params(axis='x', rotation=90)
    axs[i].set_ylim(y_min, y_max)
for j in range(len(clusters), 6):
    fig.delaxes(axs[j])
plt.tight_layout()
handles = [plt.Rectangle((0,0),1,1, color=color) for color in feature_colors]
labels = features.tolist()
fig.legend(handles, labels, loc='upper right', fontsize=12)
plt.show()
```

##### ***Group_2 :*** This group only uses the purchase method, and that also only the installment method, but the interesting part is that the full payment percentage is high, but the monthly payment is low. So, by this, we can understand the payment is done, but after a long gap. The credit card limit is very low for them, which should be increased because the full payments are done by this group.

##### ***Group_3 :*** This group uses both the features and in-purchase both the methods and has high monthly payment and the limit and balance frequency; everything is good. This group can be an ideal group and is the largest group among all others, having 25% of the total customers.

##### ***Group_4 :*** This group doesn't use any feature, 0 usage of card, but still their balance frequency is high; that should be reduced. This group consists of 0.1% of customers, so we can also neglect this group.

##### ***Group_5 :*** This group uses both the features but mostly payment in that also mostly one-off payment. Monthly payment and full payment percentage are okay, but the balance frequency is very low, which should be increased.

##### ***DBSCAN :*** This grouping shows that almost the whole data is related to each other. There are few outliers, but almost the whole data is similar, which is good; it shows how consistent the use of the card is going on.
